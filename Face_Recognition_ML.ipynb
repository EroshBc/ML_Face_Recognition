### Create data set another method

```python
labels = [i for i in os.listdir(data_path)
         if not i.startswith('.') and os.path.isdir(os.path.join(data_path,i))]

```

```python
print (labels)
```

```python
category = np.arange(len(labels))
print(categories)
```

```python
# attached categories to labels

category_dict = dict(zip(labels,categories))
print(category_dict)
```

```python
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

data=[]
target=[]


for label in labels:
    imgs_path = os.path.join(data_path, label) ## add image data path to labels
    
    img_names = os.listdir(imgs_path) # get imgaes on each directory
    
    for img_name in img_names:
        
        img_path=os.path.join(imgs_path,img_name)
        #print( img_path )
        
        img = cv2.imread(img_path)
        if img is None:
           # print("Skipping invalid image: ", img_path)
            continue
        
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces=face_classifier.detectMultiScale(gray)
        
        try:
            
        
            for (x,y,w,h) in faces:

                cropped_face=gray[y:y+h,x:x+w]
                cv2.imshow('cropped_face',cropped_face)
                print('If you accept this cropped face press y or else n:')

                k=cv2.waitKey(0) & 0xFF

                if (k == 121):
                    cropped_face=cv2.resize(cropped_face,(50,50))
                    data.append(cropped_face)
                    target.append(category_dict[label])

                else:
                    pass
        except:
            
            pass


cv2.destroyAllWindows()
sys.exit()
```
